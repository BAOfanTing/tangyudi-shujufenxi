```python
data = pd.DataFrame({'group':['a1','a2','a3','b1','b2','b3','c1','c2','c3'],
                   'data':[1,2,3,4,5,6,7,8,9]})
data
```

```python
def food_map(series):
    if series['group'] == 'a1':
        return 'A'
    elif series['group'] == 'a2':
        return 'A'
    elif series['group'] == 'a3':
        return 'A'
    elif series['group'] == 'b1':
        return 'B'
    elif series['group'] == 'b2':
        return 'B'
    elif series['group'] == 'b3':
        return 'B'
    elif series['group'] == 'c1':
        return 'C'
    elif series['group'] == 'c2':
        return 'C'
    elif series['group'] == 'c3':
        return 'C'
#要指定一个函数    
data['food_map'] = data.apply(food_map,axis = 'columns')
data
```

```python
food2Upper ={
    'A1':'a',
    'A2':'a',
    'B1':'b',
    'B2':'b',
    'C1':'c',
    'C2':'c'   
}
#转换要指定一个字典    
data['upper'] = data['group'].map(food2Upper)
data
```

### 划分区间

```python
ages = [15,18,20,21,22,34,41,52,63,79]
bins = [10,40,80]#确定划分区间
bins_res = pd.cut(ages,bins)
bins_res
```

```python
pd.value_counts(bins_res)
```

```python
pd.cut(ages,[10,30,50,81])
```

```python
group_names = ['youth','mid','old']
pd.value_counts(pd.cut(ages,[10,20,30,40],labels = group_names))
```
